```python
#! usr/bin/env python3

import os
from survey_coll import survey_output
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer

home_dir = '/media/james/Data/lsu_projects/cdm_xporter/one_off_scripts'
```

```python
coll_fields_dict = dict()

for k, v in survey_output.items():
    slim_k = k.replace(' ', '')
    for value_item in v:
        if value_item in coll_fields_dict.keys():
            coll_fields_dict[value_item] += '{}{}'.format(' ', slim_k)
        else:
            coll_fields_dict[value_item] = slim_k
```

```python
# print(coll_fields_dict)
```

```python
if 'coll_tags_as_files' not in os.listdir(home_dir):
    os.mkdir('{}/coll_tags_as_files/'.format(home_dir))
for k,v in coll_fields_dict.items():
    name = '{}/coll_tags_as_files/{}.txt'.format(home_dir, k)
    with open(name, 'w') as f:
        f.write(v)
```

```python
filenames = [name for name in os.listdir('{}/coll_tags_as_files/'.format(home_dir))]
# print(filenames)
```

```python
vectorizer = CountVectorizer(input='filename')

os.chdir('{}/coll_tags_as_files/'.format(home_dir))
dtm = vectorizer.fit_transform(filenames)  # a sparse matrix

vocab = vectorizer.get_feature_names()  # a list
```

```python
# print(dir(dtm), dtm)
print(type(dtm))
```

```python
dtm = dtm.toarray()  # convert to a regular array

vocab = np.array(vocab)
```

```python
# print(list(vocab))
# for n, i in enumerate(vocab):
    # print(dtm[0, n])
```

```python
from sklearn.metrics.pairwise import cosine_similarity

dist = 1 - cosine_similarity(dtm)
```

```python
from mpl_toolkits.mplot3d import Axes3D

mds = MDS(n_components=3, dissimilarity="precomputed", random_state=1)
pos = mds.fit_transform(dist)
fig = plt.figure(figsize=(10, 10), dpi=None, facecolor='w', edgecolor='k')
ax = fig.add_subplot(111, projection='3d')
ax.scatter(pos[:, 0], pos[:, 1], pos[:, 2])
# for x, y, z, s in zip(pos[:, 0], pos[:, 1], pos[:, 2], names):
#     ax.text(x, y, z, s)
plt.tight_layout()
plt.show()
```

```python
from scipy.cluster.hierarchy import ward, dendrogram

linkage_matrix = ward(dist)

# match dendrogram to that returned by R's hclust()
dendrogram(linkage_matrix, orientation="top", labels=names, leaf_font_size='2')


plt.tight_layout()  # fixes margins

# plt.show()
```
